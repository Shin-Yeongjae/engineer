# 4. Function show()

- show function을 사용하게 되면, 팝업창에 그래프를 보여주면서 프로그램을 멈춰줌
- 창을 닫기 전까지는 다른 프로그램이 더 이상 실행되지 않음


- 위와 같은 특성을 갖춘 함수를 Blocking Function이라고 함
    - 대표적인 함수에는 input()과 show()가 있음

```python
# 5. Extra features to decorate graphs

from pylab import plot, ylim, xlabel, ylabel, show, title
from numpy import linspace, sin

x = linspace(0, 10, 100)
y = sin(x)
y1 = cos(x)
y2 = tan(x)
plot(x, y, x, y1)
ylim(-1.1, 1.1) # y축의 범위가 -1.0 ~ 1.0이 되 것
xlabel("x axis")
ylabel("y = sin(x)")
title("cos & sin graph")
show()
```

- plot(x값, y값, 쓰고 싶은 option)
    - plot 안에 symbol을 사용할 수 있는 option이 있음
    - symbol option의 구성은
        - 첫번째 글자 : Color, 색의 첫번째 알파벳
            - 예를 들어, green이면 "g" / blue면 "b"
        - 두번째 글자 : Style
            - 예를 들어, solid line을 사용하고 싶으면 "-"
    - Color - red, green, blue, cyan, magenta, yellow
    - Style
        - "-" solid line
        - "--" dashed line
        - "o" mark points with a circle (but not connect them with lines)
        - "s" mark points with a square
        - "d" mark points with a diamond

```python
# plot option 실습

from pylab import plot, ylim, xlabel, ylabel, show, title
from numpy import linspace, sin

x = linspace(0, 10, 100)
y = sin(x)
y1 = cos(x)
y2 = tan(x)

plot(x, y1, "ko")
plot(x, y, "g--")
show()
```

# 6. Scatter plots

- 산포도를 그리는 것
- 2개의 quantities를 각각의 axis로 한 다음 점으로 찍는 것

```python
from pylab import scatter, xlabel, ylabel, xlim, ylim, show
from numpy import loadtxt

data = loadtxt("stars.txt", float)
x = data[:, 0]
y = data[:, 1]

# plot(data[:, 0], data[:, 1], "ko")

scatter(x, y)
xlabel("Temperature")
ylabel("Magnitude")
xlim(0, 13000)
ylim(-5, 20)
show()
```

# 7. Density plots : imshow (heat plot)

- 2차원 평면에 어떠한 값이 있을 때, 그 값이 많으면 진하게, 적으면 연하게 이런 식으로 하는 것

```python
from pylab import imshow, show
from numpy import loadtxt

data = loadtxt("circular.txt", float)
imshow(data)
show()
```

# 8. Function imshow

- 기본적으로 쓰는 color skin들이 여러 가지가 있음
- default color - "viridis" heat maps
    - jet, gray, hot, spectral, bone, hsv, viridis
    - redblue, redwhiteblue, inversegray
- 사용할때는 그냥,
    - jet() 로 써주면 됨.
    
    
- from pylab import imshow, show, jet, hot
- hot()
- jet()


- colorbar()
    - 오른쪽에 colorbar가 생김
    
- 특이하게 좌측 상단을 (0, 0)으로 찍는 것이 defalut 값
    - 바꾸어주려면,
        - imshow(data, origin = "lower")

- 크기를 줄이거나 늘리려면 aspect ratio option을 사용하면됨.
    - imshow(data, aspect = 2.0)

```python
from pylab import imshow, show
from numpy import loadtxt

data = loadtxt("circular.txt", float)
imshow(data, origin = "lower", aspect = 2.0)
show()
```
